```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()

# Face
face = plt.Circle((0, 0), 1, color='yellow', ec='black')
ax.add_patch(face)

# Eyes
left_eye = plt.Circle((-0.3, 0.4), 0.15, color='white', ec='black')
right_eye = plt.Circle((0.3, 0.4), 0.15, color='white', ec='black')
ax.add_patch(left_eye)
ax.add_patch(right_eye)

# Pupils
left_pupil = plt.Circle((-0.3, 0.45), 0.05, color='black')
right_pupil = plt.Circle((0.3, 0.45), 0.05, color='black')
ax.add_patch(left_pupil)
ax.add_patch(right_pupil)

# Smile
x = np.linspace(-0.4, 0.4, 100)
y = -0.5 * np.sin(2 * np.pi * x)
ax.plot(x, y, 'k', linewidth=3)

# Settings
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_aspect(1)
ax.axis("off")

plt.show()
```
