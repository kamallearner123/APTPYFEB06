```python
'''
- No space
- No specialy chars
- Only a-z A-Z 0-9 _
'''
def get_fact(num):
    result = 1
    for i in range(1,num):
        result = result*i
        print(result)
    return result


result = get_fact(10)
print(f"Factorial of number {10}, is {result}")
```

```python
'''
1) Positional Arguments (arg1, arg2)
2) Variable numer of arguments (*args)
3) Keyword arguments (**kwargs)
4) Default arguments
'''
# int add(int a, int b) {}
# add('a',20) # Warning
#1) Positional Arguments
def math(op, arg1, arg2): # int, float, True, None
    if op == "add":
        return arg1+arg2
    elif op == "sub":
        return arg1-arg2
    elif op == "mul":
        return arg1*arg2
    else:
        print("invalid operation")
        return None

#print(math("mul", "Hello ", 10.1)) # 
print(math(20, 10, "mul")) # Error : Position is not proper
#print(math("mul", "Hello ", 10))
#print(math("div", 10, 10))
```

```python
# 2) Variables num of args
def sum_all(*args):
    result = 0
    for num in args:
        result = result + num
    return result

result = sum_all(1,2,3,4)
print(result)

result = sum_all(1,2,3,4,5,6)
print(result)

result = sum_all(1,2)
print(result)
```

```python
# 2.1) Variables num of args along with Positional args
def sum_all(op, *args):
    if op == "add":
        result = 0
        for num in args:
            result = result + num
        return result
    elif op == "mul": 
        result = 1
        for num in args:
            result = result*num
        return result

result = sum_all("add", 1,2,3,4)
print(result)

result = sum_all("mul", 1,2,3,4,5,6)
print(result)

result = sum_all("add", 1,2)
print(result)
```

```python
# 4) Default arguments
def pass_port_form(name, age, place="Bangalore", citizen="Indian"):
    final_string = f"Name: {name}\nAge: {age}\nPlace: {place}\nNationality: {citizen}"
    return final_string

to_print = pass_port_form("Ram", 23)
print(to_print)

to_print = pass_port_form("Ram", 23, "Hyd", "Pak")
print(to_print)
```
